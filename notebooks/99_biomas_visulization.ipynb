```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
```

```python
SAMPLE_SIZE = 10
```

```python
file_path = "../../data/forest/scanfi/SCANFI_att_biomass_SW_2020_v1.2.tif"
```

```python
with rasterio.open(file_path) as dataset:
    print(f"CRS: {dataset.crs}")
    # Read the image data
    image_data = dataset.read(1)
    del dataset
    print(f"original size: {image_data.shape}")
```

```python
null_value = image_data.min()
```

```python
image_data[image_data == null_value] = 0
```

```python
fig, ax = plt.subplots(
    figsize = (20,20)
)

plt.imshow(
    image_data[::SAMPLE_SIZE, ::SAMPLE_SIZE], 
    cmap='inferno',
)

plt.colorbar(
    orientation='horizontal',
    pad=0.05
)
```
